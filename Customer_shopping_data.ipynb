```python
import pandas as pd
```

```python
dataset = pd.read_csv('customer_shopping_behavior.csv')
dataset.head(4)
```

```python
dataset.info()
```

```python
dataset.describe(include='all')
```

```python
dataset.isnull().sum()
```

```python
dataset['Review Rating'] = dataset.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
dataset.isnull().sum()
```

```python
dataset.columns = dataset.columns.str.lower()
dataset.columns = dataset.columns.str.replace(' ', '_') 
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
dataset.isnull().sum()  
```

```python
# Create a column Age Group based on Age
labels = [ 'Young Adult', 'Adult','Middle-aged', 'Senior']
dataset['age_group'] = pd.cut(dataset['age'], bins=[0, 25, 40, 60, 100], labels=labels)
```

```python
dataset[['age', 'age_group']].head(10)
```

```python
# Create columns pruchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Every 3 months': 90
}
dataset['purchase_frequency_days'] = dataset['frequency_of_purchases'].map(frequency_mapping)
```

```python
dataset[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```
